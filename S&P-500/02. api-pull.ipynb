```python
import numpy as np
import pandas as pd
import requests
import json
import time
```

```python
df = pd.read_csv('Stock-Background.csv')
```

```python
df = df[['Symbol', 'Security', 'GICS Sector']]
```

```python
ticker_list = np.array(df[['Symbol']]).tolist()
```

```python
print(ticker_list)
```

```python
# S&P 500 doesn't have any special characters in their stock-list, so no need for cleaning right now.
```

```python
ticker_list_clean = [i[0] for i in ticker_list]
```

```python
ticker_list_clean
```

```python
def get_ticker_json(ticker_name):
    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol={ticker_name}&apikey=[KEYHERE]"
    r = requests.get(url)
    data = r.json()
    with open("./S&P-data/" + str(ticker_name) + ".json", 'w') as f:
        json.dump(data, f)
```

```python
url = "https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol=MMM&apikey=[KEYHERE]"
```

```python
for i in ticker_list_clean:
    print(i)
    get_ticker_json(i)
    time.sleep(12.5)
```
