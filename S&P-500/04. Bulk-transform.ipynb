```python
p = Path(r'D:\Documents\Datasets\001. ML Projects\Sticker\S&P-500\S&P-data').glob('*.json')
```

```python
file_names = []
for i in p:
    file_names.append(i)
```

```python
def json_to_df(file):
    data_dict = {}
    with open(file, 'r') as f:
        
        data = json.load(f)
        name = data['Meta Data']['2. Symbol']
        for i in data['Weekly Adjusted Time Series']:
            week = str(i)

            close = str(data['Weekly Adjusted Time Series'][i]['5. adjusted close'])
            data_dict[week] = close
        
        stock_df = pd.DataFrame.from_dict(data_dict, orient='index')
        stock_df = stock_df.rename(columns={0: name})
    f.close()
    return stock_df
```
