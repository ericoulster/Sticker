```python
import pandas as pd
from scipy.stats import zscore
```

```python
df = pd.read_csv('cleaned-stocklist.csv')
```

```python
df = df[1043:].reset_index().drop(columns='index')
```

```python
df[1:].sum(axis=1)
```

```python
df[1:].count(axis=1)
```

```python
df['mkt_avg'] = df.sum(axis=1) / (df.count(axis=1) - 1)
```

```python
df
```

```python
diff_df = df.iloc[:, 1:].diff()
```

```python
diff_df['Date'] = df['Date']
```

```python
diff_df
```

```python
normed_vals = [diff_df[i] / diff_df['mkt_avg'] for i in diff_df.columns[:-2]]
```

```python
normed_df = pd.DataFrame(normed_vals)
```

```python
normed_df = normed_df.T
normed_df['mkt_avg'] = diff_df['mkt_avg']
normed_df['Date'] = diff_df['Date']
```

```python
normed_df.columns = diff_df.columns
```

```python
normed_df.to_csv('stock-differences.csv')
```
