```python
import pandas as pd
from itertools import permutations
from scipy.stats import spearmanr
import json
```

```python
df = pd.read_csv('../stock-differences.csv')
```

```python
df
```

```python
df = df.drop(columns=['Unnamed: 0', 'mkt_avg', 'Date'])
```

```python
df.iloc[:, 0].name
```

```python
df.columns = [x.split('.')[0] for x in df.columns]
```

```python
x = spearmanr(df.iloc[:, 0], df.iloc[:,1], nan_policy='omit')
y = df.iloc[:, 1].name
```

```python
for x in permutations(df.columns, 2):
    print(x[0], x[1])
```

```python


for y in df.columns:
    val_dict = {}
    for x in df.columns:
        if x != y:
            subset = df[[y, x]]
            if subset.iloc[:, 0].name == subset.iloc[:, 1].name:
                pass
            else:
                try:
                    x = spearmanr(subset.iloc[:, 0], subset.iloc[:,1], nan_policy='omit')
                    val_dict[str(y) + 'vs' + str(subset.iloc[:,1].name)] = {'cor_coef': x[0], 'p-val':x[1], 'cor_processed': 1}
                except:
                    val_dict[str(subset.iloc[:,1].name)] = {'cor_processed': 0}
        else:
            pass
    name_dict = {str(y) : val_dict}
    with open('../data/' + str(y) + '.json', 'w', encoding='utf-8') as f:
        json.dump(name_dict, f)
    del val_dict

```
